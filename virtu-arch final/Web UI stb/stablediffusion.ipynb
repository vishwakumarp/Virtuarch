```python
from diffusers import StableDiffusionImg2ImgPipeline


```

```python
model_id_or_path = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionImg2ImgPipeline.from_pretrained(model_id_or_path, torch_dtype=torch.float16)
pipe = pipe.to(device)
```

```python
init_image = Image.open("C:/Users/vishw/OneDrive/Desktop/Web UI stb/uploads/interior_design2.jpg").convert("RGB")
plt.imshow(init_image)
plt.axis('off')  # Hide the axis
plt.show()
# Resize the image to the desired size
#init_image = init_image.resize((768, 512))

# Set the prompt
prompt = "give the image a taj style base but keep the core elements intacts, just apply the style with originality preserved, give it a higher resolution render- enhance it"

# Generate the image
images = pipe(prompt=prompt, image=init_image, strength=0.40, guidance_scale=7.90,num_inference_steps=50,scheduler='K-EULER').images

# Save the image
#images[0].save("experiment.png")
image = images[0]
print((images))
```
